```python
%config IPCompleter.greedy=True
import retro
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
retro.data.list_games()
```

```python
env = retro.make(game='SuperMarioBros-Nes')
```

```python
obs = env.reset()

for i in range(50):
    random_action = env.action_space.sample() #pick a random action
    
    obs, reward, done, info = env.step(random_action) #step takes the supplied action, does it, and returns the results
    env.render()
    if done:
        obs = env.reset() #if done (game over) then reset the environment to the starting state
    plt.imshow(obs) 
env.close()
```

```python
def main():
    env = retro.make(game='SuperMarioBros-Nes')
    
    for episode in range(5):
        obs = env.reset()
        print("starting episode ", episode)
        while True:
            obs, rew, done, info = env.step(env.action_space.sample())
            env.render(mode='close')
            if done:
                print("reached done on iteration ")
                break;
                
        print("finished episode ", episode)
    env.close()


if __name__ == "__main__":
    main()
```
